```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
model.predict([[1,3,3,22,123,11,56,43,222,777,23,11,34,56,32,23,67,43,22,88,33,22,23,33,33]])
```

```python
df1=pd.DataFrame({'Actualvalue':y_test,'Predictedvalue':y_pred,'Diff':y_test-y_pred})
df1
```

```python
print("Intercept is", model.intercept_)
print("slope is", model.coef_)
```

```python
# To find accuracy, Here we are using r2_score
from sklearn.metrics import r2_score
print(r2_score(y_test,y_pred))
```
